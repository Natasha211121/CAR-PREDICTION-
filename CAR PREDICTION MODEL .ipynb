```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['VIF_FACTOR']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```
